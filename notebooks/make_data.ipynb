```python
import pandas as pd
import numpy as np
import scipy.stats
```

```python
data = pd.read_csv('../data/sub.csv')
```

```python
data.isnull().sum()
```

```python
data = pd.get_dummies(data, columns=['category_id'])
```

```python
sub = []
for s in data['comment_count']:
    if int(s) >= 150:
        sub.append(s*100)
    else:
        sub.append(s*10)
data['subscribeCount'] = sub
```

```python
data['subscribeCount']
```

```python
data.columns
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.model_selection import train_test_split
```

```python
data_x = data.loc[:, ['category_id_1', 'category_id_2', 'category_id_10', 'category_id_15', 'category_id_17', 'category_id_19', 'category_id_20', 'category_id_22', 'category_id_23', 'category_id_24', 'category_id_25', 'category_id_26', 'category_id_27', 'category_id_28', 'category_id_29', 'subscribeCount', 'sentence_importance']]
data_y = data.loc[:, ['views', 'likes']]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(data_x, data_y, train_size=0.7, random_state=42)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
import pickle
lg = Ridge()
lg.fit(X_train, Y_train)
pred = lg.predict(X_test)
print(pred)
print(lg.score(X_test, Y_test))
pickle.dump(lg, open('liner_model.sav', 'wb'))
```
